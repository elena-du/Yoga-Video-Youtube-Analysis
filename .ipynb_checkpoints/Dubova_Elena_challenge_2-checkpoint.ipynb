## Challenge 1

Plot domestic total gross over time

```python
data['ReleaseDate'] = pd.to_datetime(data['ReleaseDate'], format='%Y-%m-%d %H:%M:%S')
```

```python
data.sort_values('ReleaseDate', inplace = True)
```

```python
plt.figure(figsize = [12, 4])
plt.plot(data['ReleaseDate'], data['DomesticTotalGross'])
plt.title('Domestic Total Gross over Release Date', fontsize=20, loc='left', pad=15)
plt.xlabel('Release Date', labelpad = 10)
plt.ylabel('Domestic Total Gross', labelpad = 10);
```

## Challenge 2

Plot runtime vs. domestic total gross.

## Challenge 4

Make one figure with (N=numbber of MPAA ratings there are) subplots, and in each plot the release date vs the domestic total gross.

```python
data['Rating'].unique()
```

```python
data_R = data[data['Rating'] == 'R']
data_PG_13 = data[data['Rating'] == 'PG-13']
data_PG = data[data['Rating'] == 'PG']
data_G = data[data['Rating'] == 'G']
```

```python
fig,ax = plt.subplots(2,2,sharex=True,sharey=True,figsize=[12,7])
plt.suptitle('Release Date vs Domestic Total Gross by Rating', fontsize=16)

# Row 1 

ax[0][0].plot(data_R['ReleaseDate'], data_R['DomesticTotalGross'])
ax[0][0].set_title('R')
ax[0][1].plot(data_G['ReleaseDate'], data_G['DomesticTotalGross'])
ax[0][1].set_title('G')
# Row 2 

ax[1][0].plot(data_PG['ReleaseDate'], data_PG['DomesticTotalGross'])
ax[1][0].set_title('PG')
ax[1][1].plot(data_PG_13['ReleaseDate'], data_PG_13['DomesticTotalGross'])
ax[1][1].set_title('PG_13');
```

## Challenge 6

Bin your dataset into months and make a bar graph of the mean domestic total gross by month. Error bars will represent the standard error of the mean. Title of the graph should include: Mean Domestic Total Gross by Month in 2013. 

```python
data.index = pd.to_datetime(data['ReleaseDate'],format='%Y/%m/%d')
```

```python
data_by_month = data.groupby([data.index.month])['DomesticTotalGross'].mean()
data_by_month = pd.DataFrame(data_by_month)
data_by_month
```

```python
ax = data_by_month[['DomesticTotalGross']].plot(kind='bar', title ="Mean Domestic Total Gross by Month in 2013", figsize=(12, 7), legend=True, fontsize=12)
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Total Domestic Gross (100Ms)", fontsize=12);
```
